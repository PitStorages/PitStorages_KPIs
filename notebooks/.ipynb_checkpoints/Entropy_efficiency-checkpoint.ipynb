# Entropy efficiency

This script performs a simple calculation of the entropy efficiency for a 1-D storage in order to provide a deeper understaning on how to assess stratification of a storage using the entropy efficiency.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from temperature_profile import temp_profile
from matplotlib import animation
from matplotlib.animation import FuncAnimation
```

Enter the simulation settings.

By default the ambient temperature is set to $10 ^oC$ and the time-step duration to $1$.

The value for the heat loss coefficient (850) cannot be in W/K, this value is way too high! It has to be in J/K. This way, if it is converted to W/K it corresponds to 0.24 W/K which is a much more reasonalble value for a tank.

```python
# Water proprties
density_water = 980 # in kg/m3
specific_heat_water = 4200  # in J/(kg K)

# Simulation properties
simulated_case = '' # Set the case that will be simulated, e.g. "fully_mixed", "fully_stratified". Otherwise, it simulates storage operation with mixing based on mixing_nodes.

N = 5000 # number of tank nodes
mixing_nodes = 100 # gives the number of nodes that will be mixed (average temperature) at each timestep.

T_hot = 90 # charge temperature
T_cold = 45 # discharge temperature
T_threshold = 10 # temperature threshold for considering the tank full/empty. E.g. if it is 10 degC then the tank is fully when the bottom is at 80 degC

charge_time = 7*24  # number of timesteps used for charging
discharge_time = 7*24  # number ofthimesteps used for discharging
simulation_period = 1500 # in hours

heat_loss_coeff = 850 # in J/(m^2 K). This number gives a 90% storage efficiency for the selected settings. Set to 0 for no heat loss.
T_amb = 10 # ambient temperatrue for calculation of heat losses

max_storage_cycles = 2 # stops the simulation after a number of charge-discharge cyles

# Plots the temperature profile in the tank
plot = False
```

Create dataframe containing the temperature profiles for each time step.Rows are the simulation time steps and columns are the layers of the storage starting from the top (0) to the bottom (N).

The "df_charge_status" is an indication when the storage is charging (1) or discharging (0).

```python
df, df_ancil = temp_profile(simulated_case, N, mixing_nodes, T_hot, T_cold, T_threshold,
                             simulation_period, charge_time, discharge_time, heat_loss_coeff,T_amb = T_amb, max_storage_cycles=max_storage_cycles, plot=plot)
```

It is probalby not correct to use the entropy function (from engineering toolbox) for finding the entropy of the storage. The reason is that this function assumes zero entropy for water at 0 $^oC$ but the storage is considered empty when the temperature is 45 $^oC$. Although from a physics point of view this is correct (water has entropy in temperatures higher than $0 ^oC$), regarding the storage operation it gives a paradox, since the storage is considered empty regarding energy but still has an entropy content! 

It is probalby better to calculate the entropy of the storage using the expression: $$\Delta S = \frac{\Delta Q}{T}$$

This way there is an agreement between the storage energy and entropy content.

The same method should be followed for the entropy flows in/out of the storage. Haller suggests to calculate them using a standard entropy equation. However, this way, if the storage is at $50 ^oC$ and the inlet temperature is at the same temperatue, there will be higher entropy in the flow thant in the storage. For this reason, also for the flows, the above formula is recommended.

In the end, all of the storage's entropies (storage, flows and heat loss) will be calculated using the same formula so at least the calculation will be consistent!

Then for finding the entropy efficiency, the internally generated entropy in the storage has to be divided with the internally generated entropy of a reference storage that is considered fully mixed during the entire charge/discharge operation:

$$\eta_{S} = 1 - \frac{\Delta S_{exp,int}}{\Delta S_{mix,int}}$$

Where:
$$\Delta S_{int} = \Delta S_{storage} - \Delta S_{inlet} + \Delta S_{outlet} + \Delta S_{heat loss}$$

This entropy balance is derrived from the general expression:
$$S_{input} + S_{generation} - S_{output} = S_{accumulation}$$

Considerations:
* The mixed storage entropy should be higher than the stratified.
* Since the storage is an open system (heat and mass can flow through it) the storage entropy can be increased and decreased  (but cannot be negative).
* Both the $S_{exp,int}$ and the $S_{mix,int}$ should be positive since it is entropy generated by mixing, heat conduction and heat diffusion.

```python
storage_height = 1 # the height of storage in m
storage_volume = 1 # the volume of storage in m^3
volume_per_layer = 1/N # assuming that the storage is 1-D [m^3]
mass_per_layer = density_water * volume_per_layer # mass per layer [kg]
storage_mass = density_water * storage_volume # total storage mass [kg]

# weighted average temparature of the storage at each timestep
T_avg = (df*mass_per_layer/storage_mass).sum(axis=1)
```

```python
# For the simulated storage
_ = df[np.arange(0,df.shape[1],400)].plot(cmap='rainbow_r', alpha=0.5, figsize=(12,5))
```

Calculation of entropies for the experimental storage. First the inlet and outlet temperatures are defined and then entropies are calculated for the flow, storage, heat loss and internally generated.

```python
time_step = 1 # [h] this number should not be changed in order to avoid numerical diffusion!

charge_nodes = np.max([int(np.round(N/(charge_time/time_step))), 1]) # number of nodes charged each time step (avoid numerical diffusion)
discharge_nodes = np.max([int(np.round(N/(discharge_time/time_step))), 1]) # number of nodes discharged each time step (to avoid numerical diffusion)

# Flows in [m3]
flow_in = charge_nodes * volume_per_layer
flow_out = discharge_nodes * volume_per_layer

T_input = []
T_output = []

for i in df.index:
    if df_ancil['charge_status'][i]==1:
        T_i = df.loc[i,0]
        T_o = df.loc[i,N-1]
        T_input.append(T_i)
        T_output.append(T_o)
    else:
        T_i = df.loc[i,N-1]
        T_o = df.loc[i,0]
        T_input.append(T_i)
        T_output.append(T_o)

T_in = pd.Series(T_input)
T_out = pd.Series(T_output)


# Entropy in and out of the storage [J/K]
S_exp_in = (flow_in * density_water * specific_heat_water * (T_in - T_cold))/(T_in + 273.15)
S_exp_out = (flow_out * density_water * specific_heat_water * (T_out - T_cold))/(T_out + 273.15)

# The calculated Q_store is identical to the df_ancil['energy_content'], units in [J/K]
S_exp_accumulated = (storage_volume * density_water * specific_heat_water * (T_avg - T_cold))/(T_avg+273.15)

# The entropy flow and heat loss are time dependent with a time step of 1, so this is why cumsum is used
S_exp_flow = (S_exp_in-S_exp_out).cumsum() # [J/K]
S_exp_heat_loss = -(heat_loss_coeff * (T_avg - T_amb) / (T_avg+273.15)).cumsum() # [J/K]

S_exp_internal = S_exp_accumulated - S_exp_flow - S_exp_heat_loss

plt.figure(figsize=(12,5))
S_exp_internal.plot(label='internal')
S_exp_flow.plot(label='flow')
S_exp_accumulated.plot(label='accumulated')
S_exp_heat_loss.plot(label='heat loss')
plt.legend()
plt.axhline(0, c='k')
```

For time periods with constant charging or discharging mass flow and inlet temperature, the temperature of the fully mixed reference storage at time $t_{B}$ is calculated from the temperature at the begining of this period $T_{store,mix}(t_{A})$, the temperature that would be reached after infinite time $T_{inf}$, and the time constant of the process $a$.

$$ T_{out,mix}(t_{B}) = T_{inf} + [T_{store,mix}(t_{A}) - T_{inf}] \cdot exp(-a \cdot [t_{B}-t_{A}]) $$

With:

$$ T_{inf} = \frac{UA \cdot T_{amb} + C_{flow} \cdot T_{inlet}}{UA + C_{flow}} $$

$$ a = \frac{UA + C_{flow}}{C_{store}} $$

Where $C_{flow}$ is the capacity flow rate of the charging or discharging process and $C_{store}$ is the thermal capacitance of the storage.

$$ C_{flow} = \frac{H(T_{out}) - H(T_{inlet})}{T_{out} - T_{inlet}} $$

Where $h$ is the enthalpy in $J$.

The fully mixed reference storage is assumed to be equal to the experimental at the begining of the experiment and fully mixed from this point until the end of the experiment. Further, the fully mixed reference storage has the same heat loss coefficients as the experimental storage and is subjected to the same mass flows and temperatures. The temperature of the fully mixed reference storage is at the same time its outlet temperature.

```python
# Enthalpy (energy) in and out of the storage [J]
H_in = flow_in * density_water * specific_heat_water * (T_in - T_cold)
H_out = flow_out * density_water * specific_heat_water * (T_out - T_cold)

# Capacity flow rate of charge/discharge [J/K]
C_flow = (H_out - H_in).divide(T_out - T_in)

# Thermal capacitance of storage [J/K]
C_store = specific_heat_water * storage_mass

# Time constant [s^-1]. Heat loss coefficient has to be converted to [J/(s K)] in order for the time constant to have units of [s^-1]
a = (heat_loss_coeff/3600 + C_flow)/C_store

# Temperature of storage that would be reached after infinite time [C]
T_inf = ((heat_loss_coeff*(T_amb+273.15) + C_flow*(T_in+273.15))/(heat_loss_coeff + C_flow))-273.15

# Outlet temperature of the fully mixed storage [C]
T_out_mix = []

# list with the timesteps where the 'charge_status' changes
change_operation = df_ancil['charge_status'][df_ancil['charge_status'].diff() != 0].index.tolist()

# Empty list for adding the outlet temperatures for the fully mixed storage
T_out_m = []

# Iterative calculation for the outlet temperature of the fully mixed storage [C]
for i in df.index:
    if i in change_operation:
        pos = i
        
    if i < change_operation[1]:
        # the fully mixed storage starts with the same temperature profile as the actual storage
        T_o_m = T_inf[i] + ((df.loc[0]*mass_per_layer/storage_mass).sum() - T_inf[i]) * np.exp(-a*(i-pos))[i]
        T_out_m.append(T_o_m)
    else:
        T_o_m = T_inf[i] + (T_out_m[pos-1] - T_inf[i]) * np.exp(-a*(i-pos))[i]
        T_out_m.append(T_o_m)
            
T_out_mix = pd.Series(T_out_m)

# The temperature of the mixed storage is equal to the outlet temperature [C]
T_store_mix = T_out_mix
```

Inspection of the calculated mixed storage output temperature compared to the average temperatrue of the experimental storage and the inifinite temperatrue of the fully mixed storage.

```python
plt.figure()
T_out_mix.plot(label='T_out fully mixed storage')
T_avg.plot(label='T_avg experimetal storage')
T_inf.plot(label='T_inf fully mixed storage')
plt.legend(bbox_to_anchor=(1,1))
```

Entropy calculation for the fully mixed storage

```python
S_mix_in = (flow_in * density_water * specific_heat_water * (T_in - T_cold))/(T_in + 273.15)
S_mix_out = (flow_out * density_water * specific_heat_water * (T_out_mix - T_cold))/(T_out_mix + 273.15)

# The calculated Q_store is identical to the df_ancil['energy_content']
S_mix_accumulated = (storage_volume * density_water * specific_heat_water * (T_store_mix - T_cold))/(T_store_mix+273.15)

# The entropy flow and heat loss are time dependent so this is why sumsum is used
S_mix_flow = (S_mix_in-S_mix_out).cumsum()
S_mix_heat_loss = -(heat_loss_coeff * (T_store_mix - T_amb) / (T_store_mix+273.15)).cumsum()

S_mix_internal = S_mix_accumulated - S_mix_flow - S_mix_heat_loss

plt.figure(figsize=(12,5))
S_mix_internal.plot(label='internal')
S_mix_flow.plot(label='flow')
S_mix_accumulated.plot(label='accumulated')
S_mix_heat_loss.plot(label='heat loss')
plt.legend()
plt.axhline(0, c='k')
```

Entropy efficiency calculation

```python
S_eff = (1 - S_exp_internal/S_mix_internal)*100

plt.figure(figsize=(12,5))
S_eff.plot()
plt.ylim(0,105)
```

This result is a bit difficult to interpret. It would make more sense to have the overall entropy efficiency of the storage for its operation
$$ \eta_{S} = \frac{S_{out}}{S_{in}} $$

```python
charging_timestep = df_ancil['charge_status'][(df_ancil['charge_status'].diff() != 0) & (df_ancil['charge_status'] == 1)].index.tolist()

S_eff_cycle_1 = S_exp_out[:charging_timestep[1]].sum()/S_exp_in[:charging_timestep[1]].sum()

S_eff_cycle_2 = S_exp_out[charging_timestep[1]:].sum()/S_exp_in[charging_timestep[1]:].sum()

S_eff_overall = S_exp_out.sum()/S_exp_in.sum()

print(f'The entropy efficiency of the 1st cycle is: {round(S_eff_cycle_1,3)*100}%')
print(f'The entropy efficiency of the 2nd cycle is: {round(S_eff_cycle_2,3)*100}%')
print(f'The overall entropy efficiency is: {round(S_eff_overall,3)*100}%')
```

```python
Q_exp_in = flow_in * density_water * specific_heat_water * (T_in - T_cold)
Q_exp_out = flow_out * density_water * specific_heat_water * (T_out - T_cold)

Q_eff_overall = Q_exp_out.sum()/Q_exp_in.sum()
print(f'The overall energy efficiency is: {round(Q_eff_overall,3)*100}%')
```

Equation for finding entropy of water based on temperature (0 C has 0 entropy).
Derived from https://www.engineeringtoolbox.com/water-properties-d_1508.html

```python
def s(T):
    '''This function finds the specific entropy in J/(kg K), temperature is in C.'''
    S = -0.0000000002*T**4 + 0.0000001031*T**3 - 0.0000313286*T**2 + 0.0154172716*T - 0.0001227190
    return S*1000

def h(T):
    '''This function finds the specific enthalpy in J/kg, temperature is in C.'''
    H = 0.0000857036*T**2 + 4.1800307227*T + 0.1602492667
    return H*1000
```
