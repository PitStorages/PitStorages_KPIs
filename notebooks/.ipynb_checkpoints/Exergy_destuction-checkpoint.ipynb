```python
#%matplotlib widget
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from temperature_profile import temp_profile
from matplotlib import animation
from matplotlib.animation import FuncAnimation
```

```python
# Water proprties
density_water = 980 # in kg/m3
specific_heat_water = 4200  # in J/(kg K)

# Simulation properties
simulated_case = '' # Set the case that will be simulated, e.g. "fully_mixed", "fully_stratified". Otherwise, it simulates storage operation with mixing based on mixing_nodes.

N = 60 # number of tank nodes

T_hot = 90 # charge temperature
T_cold = 45 # discharge temperature
T_threshold = 10 # temperature threshold for considering the tank full/empty. E.g. if it is 10 degC then the tank is fully when the bottom is at 80 degC

simulation_period = 1500

heat_loss_coeff = 1850 # This number gives a 90% storage efficiency for the selected settings. Set to 0 for no heat loss.

max_storage_cycles = 2 # stops the simulation after a number of charge-discharge cyles

# Plots the temperature profile in the tank
plot = False
```

```python
df, df_ancil = temp_profile(simulated_case, N, T_hot, T_cold, T_threshold, simulation_period,
                            heat_loss_coeff, plot=plot, max_storage_cycles=max_storage_cycles, T_amb=10)
```

```python
_ = df[np.arange(0,df.shape[1],5)].plot(cmap='rainbow_r', alpha=0.5, figsize=(12,5))
```

```python
storage_height = 1 # the height of storage in m
storage_volume = 1 # the volume of storage in m^3
volume_per_layer = 1/N # assuming that the storage is 1-D
mass_per_layer = density_water * volume_per_layer # mass per layer
storage_mass = density_water * storage_volume # total storage mass
```

```python
T_0 = T_cold + 273.15 # dead state temperature
T_amb = 10

# Flows in [m3]
flow_in = volume_per_layer
flow_out = volume_per_layer

T_input = []
T_output = []

for i in df.index:
    if df_ancil['charge_status'][i]==1:
        T_i = 90
        T_o = df.loc[i,N-1]
        T_input.append(T_i)
        T_output.append(T_o)
    else:
        T_i = T_cold
        T_o = df.loc[i,0]
        T_input.append(T_i)
        T_output.append(T_o)

T_in = pd.Series(T_input)
T_out = pd.Series(T_output)


# Enthalpy in and out of the storage [J]
H_in = flow_in * density_water * specific_heat_water * (T_in - T_cold)
H_out = flow_out * density_water * specific_heat_water * (T_out - T_cold)

# Enthalpy (energy) of the storage, units in [J]
H_accumulated = df_ancil['energy_content']

H_flow = (H_in-H_out) # [J]

H_heat_loss = - df_ancil['heat_loss'] # [J]


# Entropy in and out of the storage [J/K]
S_in = (flow_in * density_water * specific_heat_water * (T_in - T_cold))/(T_in + 273.15)
S_out = (flow_out * density_water * specific_heat_water * (T_out - T_cold))/(T_out + 273.15)

# The calculated Q_store is identical to the df_ancil['energy_content'], units in [J/K]
S_accumulated = df_ancil['storage_entropy']

# The entropy flow and heat loss are time dependent with a time step of 1, so this is why cumsum is used
S_flow = (S_in-S_out) # [J/K]

S_internal_generated = df_ancil['internal_entropy_generation']
S_heat_loss = df_ancil['heat_loss_entropy']

# Exergy calculations [J]
Ex_flow = H_flow - T_0 * S_flow

Ex_accumulated = (H_accumulated - T_0 * S_accumulated).diff()

# There is no energy creation due to mixing so DeltaH = 0
Ex_destruction = - T_0 * S_internal_generated

Ex_heat_loss = H_heat_loss/60 - T_0 * S_heat_loss # IT LOOKS LIKE THERE IS SOME ISSUE WITH THE CALCULATIONS!! DIVISION WITH THE NODE NUMBER IS NECESSARY!!

Ex_heat_loss_calc = Ex_accumulated - Ex_flow - Ex_destruction
```

```python
plt.figure(figsize=(12,5))
Ex_flow.plot(label='flow')
Ex_destruction.plot(label='Exergy destruction mixing')
Ex_accumulated.plot(label='internal change', linestyle='--')
Ex_heat_loss.plot(label='Heat loss')
Ex_heat_loss_calc.plot(label='Heat loss calc')
plt.legend()
plt.axhline(0, c='k')
```

```python
T_ref = T_cold # reference temperature for calculating energy

Q_storage = (density_water * volume_per_layer * specific_heat_water * (df - T_ref)).sum(axis='columns') # energy content of storage
```

```python
Q_layer = density_water * volume_per_layer * specific_heat_water * (df - T_ref)
S_layer = (density_water * volume_per_layer * specific_heat_water * (df - T_ref))/(df+273.15)

Ex_layer = Q_layer - (T_ref+273.15) * S_layer

Ex_storage = Ex_layer.sum(axis='columns')
```

```python
V_hot = Q_storage/(density_water * specific_heat_water * (T_hot - T_ref)) # volume of hot water in the stratified storage using energy content of actual storage

S_hot = (density_water * V_hot * specific_heat_water * (T_hot - T_ref))/(T_hot+273.15)

Ex_hot = Q_storage - (T_ref+273.15) * S_hot
```

```python
T_avg = (df*mass_per_layer/storage_mass).sum(axis=1) # weighted average temperature of mixed storage

S_mix = (density_water * storage_volume * specific_heat_water * (T_avg - T_ref))/(T_avg+273.15)

Ex_mix = Q_storage - (T_ref+273.15) * S_mix
```

```python
t = (Ex_hot - Ex_storage)/(Ex_hot - Ex_mix)

t.plot()
plt.ylim(0,1.2)
```
